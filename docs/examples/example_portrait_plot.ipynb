## Add Images to the plot

```python
# Test dummy array -- fill random numbers for each array
data1 = np.random.randn(3, 3)
data2 = np.random.randn(3, 3)
```

```python
xaxis_labels = ['A1', 'A2', 'A3']
yaxis_labels = ['B1', 'B2', 'B3']

xaxis_labels = ['Metric '+x for x in xaxis_labels]
yaxis_labels = ['Model '+y for y in yaxis_labels]
```

```python
import os

directory = 'images/portrait_plot'
image_list = [os.path.join(directory, filename) for filename in os.listdir(directory) if os.path.isfile(os.path.join(directory, filename))]
images = np.array(image_list, dtype=str)

names = []
for n in np.arange(0,18):
    name = 'image' + str(n)
    names.append(name)
```

```python
p = portrait_plot(
    [data1, data2],
    xaxis_labels=xaxis_labels,
    yaxis_labels=yaxis_labels,
    legend_labels=['data1', 'data2'],
    width=600,
    images=image_list,
    names=names
)
```

## Save the plot

```python
# set output to static HTML file
output_file(filename="interactive_portrait_plot.html", title="Interactive Portrait Plot")

# save the results to a file
save(p)

# Result: [`interactive_portrait_plot.html`](interactive_scatter_plot.html)
```
