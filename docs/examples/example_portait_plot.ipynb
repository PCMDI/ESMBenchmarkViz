## 1: Basic usage

Example 1 show basic usage of the function.

```python
data_all = np.stack([data1, data2, data3, data4])
#data_all = [data1, data2, data3, data4]

xaxis_labels = ['A1', 'A2', 'A3', 'A4', 'A5', 'A6', 'A7']
yaxis_labels = ['B1', 'B2', 'B3', 'B4', 'B5', 'B6', 'B7', 'B8']

xaxis_labels = ['Metric '+x for x in xaxis_labels]
yaxis_labels = ['Model '+y for y in yaxis_labels]
```

```python
p = portrait_plot(
    data1,  # or [data1, data2, data3, data4] (top, right, bottom, left: clockwise from top)
    xaxis_labels=xaxis_labels,
    yaxis_labels=yaxis_labels,
    width=600,
    debug=True
)
```

```python
p = portrait_plot(
    [data1, data2],  # or [data1, data2, data3, data4] (top, right, bottom, left: clockwise from top)
    xaxis_labels=xaxis_labels,
    yaxis_labels=yaxis_labels,
    width=600,
    debug=True
)
```

```python
p = portrait_plot(
    data_all,  # or [data1, data2, data3, data4] (top, right, bottom, left: clockwise from top)
    xaxis_labels=xaxis_labels,
    yaxis_labels=yaxis_labels,
    legend_labels=['Data1', 'Data2', 'Data3', 'Data4'], 
    width=600,
)
```
